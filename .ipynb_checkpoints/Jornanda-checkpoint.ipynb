```python
import pandas as pd
import numpy as np
from unidecode import unidecode

import re
```

```python
df_hist_horario = pd.read_excel(r'BD\Dados Vivante Funcionários.xls', sheet_name='HISTÓRICO DE HORÁRIO')
df_hist_horario.head(20)
```

```python
df_hist_horario.shape
```

```python
df_hist_horario['HORÁRIO ATUAL'].unique()
```

```python
df_horario = df_hist_horario[df_hist_horario['HORÁRIO ATUAL'] == 'Sim']
df_horario = df_horario[['MATRÍCULA', 'HORÁRIO']]
df_horario['HORÁRIO'] = df_horario['HORÁRIO'].map(lambda x: unidecode(x.lower().replace(' ', '')))
df_horario.head()
```

```python
df_horario.shape
```

```python
df_horario['Jornada - 5x2'] = df_horario['HORÁRIO'].map(lambda x: 1 if ('5x2' in x) or ('5/2' in x) else 0)
df_horario['Jornada - 6x1'] = df_horario['HORÁRIO'].map(lambda x: 1 if '6x1' in x else 0)
df_horario['Jornada - 12x36'] = df_horario['HORÁRIO'].map(lambda x: 1 if '12x36' in x else 0)
df_horario.head(10)
```

```python
df_horario.shape
```

```python
df_horario[df_horario['Jornada - 5x2']+df_horario['Jornada - 6x1']+df_horario['Jornada - 12x36'] == 0].shape
```

```python
df_horario[df_horario['Jornada - 5x2']+df_horario['Jornada - 6x1']+df_horario['Jornada - 12x36'] == 0]
```

```python
df_horario['Jornada - movel'] = df_horario['HORÁRIO'].map(lambda x: 1 if (('móvel') in x) or (('movel') in x) else 0)
df_horario['Jornada - fixo'] = df_horario['HORÁRIO'].map(lambda x: 1 if (('fix') in x) else 0)
df_horario
```

```python
df_horario.shape
```

```python
df_horario[df_horario['Jornada - movel']+df_horario['Jornada - fixo'] == 0].shape
```

```python
df_horario[df_horario['Jornada - movel']+df_horario['Jornada - fixo'] == 0]
```
