```python
import pandas as pd
import numpy as np
```

```python
df_horas = pd.read_excel(r'BD\Horas Extras Por Dia - 01-2020 a 01-2021.xlsx', sheet_name = "Planilha1")
df_func = pd.read_excel(r'BD\Dados Vivante Funcionários.xlsx', sheet_name='DADOS FUNCIONAIS')
```

```python
df_horas.info()
```

```python
df_func.info()
```

```python
df_horas.head()
```

```python
df_horas['SO'] = df_horas['Horas Extras Pagas (SO)'].map(lambda x: (int(x[0:2])*60) + int(x[3:])) + df_horas['Horas Extras Pagas em Feriado (SO)'].map(lambda x: (int(x[0:2])*60) + int(x[3:]))
df_horas[df_horas['Horas Extras Pagas (SO)'] != '00:00'].head()
```

```python
df_horas['SX'] = df_horas['Horas Extras Pagas (SX)'].map(lambda x: (int(x[0:2])*60) + int(x[3:])) + df_horas['Horas Extras Pagas em Feriado (SX)'].map(lambda x: (int(x[0:2])*60) + int(x[3:]))
df_horas[df_horas['Horas Extras Pagas (SX)'] != '00:00'].head()
```

```python
df_horas = df_horas[['Matricula', 'Data', 'SO', 'SX']]
df_horas.head()
```

```python
df_horas = df_horas.groupby('Matricula')['SO','SX'].sum().reset_index()
df_horas.head()
```

```python
df_func.head()
```

```python
df_salario = df_func[['MATRÍCULA', 'SALÁRIO']]
df_salario.head()
```

```python
df_tempo = df_func[['MATRÍCULA', 'ADMISSÃO']]
df_tempo.head()
```

```python
today = pd.to_datetime('today')

df_tempo['Tempo de casa'] = (today - df_tempo['ADMISSÃO']) / np.timedelta64(1,'M')
df_tempo = df_tempo.drop('ADMISSÃO',axis=1)

df_tempo.head()
```

```python
df = df_tempo.merge(df_salario, how='inner', on='MATRÍCULA')
df.head()
```

```python
df = df.merge(df_horas, how = 'left', left_on='MATRÍCULA', right_on='Matricula')
df = df.drop('Matricula',axis=1)
df.head()
```

```python
df[df['SO'] > 0].head()
```

```python
df.to_excel(r'BD Tratado\analiseFunc.xlsx',index=False)
```

```python
df.to_csv(r'BD Tratado\analiseFunc.csv',sep=';',index=False)
```
