```python
import pandas as pd
```

```python
df_cargo = pd.read_excel(r'BD\Dados Vivante Funcionários.xlsx',sheet_name='DADOS FUNCIONAIS')
df_cargo.head()
```

```python
df_de_para = pd.read_csv('dePara.csv',sep=';')
df_de_para.head()
```

```python
df_cargo = df_cargo.merge(df_de_para,left_on='SINDICATO',right_on='De')
df_cargo['agr_SINDICATO'] = df_cargo['Para']
df_cargo = df_cargo.drop(['De','Para'],axis=1)
df_cargo.info()
```

```python
df_cargo = df_cargo[['MATRÍCULA', 'agr_SINDICATO']]
df_cargo.to_csv(r'BD Tratado\cargo.csv',sep = ';',index=False)
df_cargo.head()
```
